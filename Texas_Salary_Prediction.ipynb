# Domain Analysis

## Domain Analysis gives the domain knowledge of all the features in data.

## Features:

1. AGENCY- It is the Agency Code of name of agency.
2. AGENCY NAME- There are certain names of agencies which have unique code.
3. LAST NAME-  The last name of a Employee.
4. FIRST NAME- The first name of a Employee.
5. MI - Middle Initial of a name.
6. Class Code- It is the designations assigned to job titles based on job content and rankings.
7. Class Title- It is the official dseignation to which a position has been assigned.
8. Ethnincity- The dress code of Employee.
9. Gender- The chararcteristics of men or women.
10. Status- It tells us whether the person is Regular or Temporary based on Full-Time or Part-time.
11. EMPLOY DATE- The date of joining of Employee.
12. HRLY RATE- The amount of money that is paid for every hour worked.
13. HRS PER WK- The no. of hours worked per week.
14. MONTHLY- The salary of an employee on monthly basis.
15. ANNUAL- The salary of an employee on yearly basis.
16. State Number- It is a unique number given by state to organizations for hiring employees.
17. duplicated- It tells us whether the records are repeated or not.
18. multiple_full_time_jobs- Employees working Full-time for multiple companies.
19. combined_multiple_jobs- It merges multiple jobs from the same market view into a single job.
20. summed_annual_salary- The summed salary of a person working on different status.
21. hide_from _search- It gives you the option for your ##searches to be hidden with respect to organisations.

# **EDA**

## Analysing the data based on graphs and trvy to get insights from different graphs.



```python
new_data= data.drop(['AGENCY','LAST NAME','FIRST NAME','MI','CLASS_CODE','STATE_NUMBER','duplicated','multiple_full_time_jobs','combined_multiple_jobs','summed_annual_salary','hide_from_search','EMPLOY_DATE'],axis=1)
new_data
```

```python
plt.figure(figsize=(20,25),facecolor='white')
plotnumber=1

for column in new_data:
    if plotnumber<=9:
    
        ax=plt.subplot(3,3,plotnumber)
        sns.histplot(x=new_data[column],kde=True,bins=50)
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Count',fontsize=20)
    plotnumber+=1
plt.tight_layout()
```
